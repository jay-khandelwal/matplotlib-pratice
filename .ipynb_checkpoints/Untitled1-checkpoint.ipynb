```python
import matplotlib.pyplot as plt
import numpy as np

A = np.random.rand(5, 5)

fig, axs = plt.subplots(1, 3, figsize=(10, 3)) 

print('fig',fig)
print('axs',axs)
for ax, interp in zip(axs, ['nearest', 'bilinear', 'bicubic']):
    print('ax :',ax)
    ax.imshow(A, interpolation=interp)
    ax.set_title(interp.capitalize())
    ax.grid(True)
    
   

#plt.show(
```
