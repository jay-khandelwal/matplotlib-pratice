```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
google_data = pd.read_csv('49864_274957_compressed_googleplaystore.csv/googleplaystore.csv',
                          nrows=1000)
```

```python
google_data.shape
```

```python
google_data
```

```python
rating = google_data['Rating']
downloads = google_data['Installs']
reviews = google_data['Reviews']
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(16,9))
plt.scatter(rating, downloads, 
            c='green',
            label='T. downloads)
plt.scatter(rating, reviews, 
          c='blue',
       label='reviews')
plt.xlabel('Downloads and review')
plt.ylabel('Rating')
plt.xticks(rotation=20)
plt.title('google play store data')
plt.legend()
plt.show()
```
